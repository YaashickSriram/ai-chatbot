```python
# Check if ENV Vars are loaded correctly:
from dotenv import load_dotenv
load_dotenv()
import os

print(f"AZ_OAI_DEPLOYMENT:   '{os.getenv('AZ_OAI_DEPLOYMENT')}'")
print(f"AZ_OAI_API_VERSION:  '{os.getenv('AZ_OAI_API_VERSION')}'")
print(f"AZ_OAI_ENDPOINT:     '{os.getenv('AZ_OAI_ENDPOINT')}'")
print(f"AZ_OAI_API_KEY:      '{os.getenv('AZ_OAI_API_KEY')}'")
```

```python
# Establish Azure OpenAI connection using Langchain wrapper:
from langchain_openai import AzureChatOpenAI
print("Imports Successful!!")

def get_llm(max_tokens=None, retries=None):
    AZ_OAI_DEPLOYMENT=os.getenv("AZ_OAI_DEPLOYMENT")
    AZ_OAI_API_VERSION=os.getenv("AZ_OAI_API_VERSION")
    AZ_OAI_ENDPOINT=os.getenv("AZ_OAI_ENDPOINT")
    AZ_OAI_API_KEY=os.getenv("AZ_OAI_API_KEY")

    llm = AzureChatOpenAI(
        azure_deployment=AZ_OAI_DEPLOYMENT,
        openai_api_version=AZ_OAI_API_VERSION,
        openai_api_key=AZ_OAI_API_KEY,
        azure_endpoint=AZ_OAI_ENDPOINT,
        temperature=0,
        max_tokens=max_tokens,
        timeout=None,
        max_retries=retries
    )
    return llm

# Initiate azure_client:
print("Instantiating LLM")
azure_client = get_llm()
print("LLM Instantiation Successful!!")

 
```

```python
# Sample execution
prompt_text = f"""
Hi, how do LLMs work?
"""
response = azure_client.invoke(prompt_text) # LLM Call
# print(response)
print(response.content)
```
